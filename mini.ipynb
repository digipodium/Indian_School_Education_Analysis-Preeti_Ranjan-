```python

  
# creating the bar plot for primary schools

fig = plt.figure(figsize = (20,5)) 
plt.bar(x=comp['State_UT'], height= comp['Primary_Only'],color ='maroon', width = 0.4) 
plt.xlabel("STATES") 
plt.ylabel("Primary_Only") 
plt.title("Students enrolled in COMPUTER courses") 
plt.xticks(rotation=80)
plt.show() 

# creating the bar plot for upper-primary schools
fig = plt.figure(figsize = (20,5)) 
plt.bar(x=comp['State_UT'], height= comp['U_Primary_Only'],color ='maroon', width = 0.8) 
plt.xlabel("STATES") 
plt.ylabel("U_Primary_Only") 
plt.title("Students enrolled in COMPUTER courses") 
plt.xticks(rotation=80)
plt.show()

# creating the bar plot for secondary schools
fig = plt.figure(figsize = (20, 5)) 
plt.bar(x=comp['State_UT'], height= comp['Sec_Only'],color ='maroon', width = 0.8) 
plt.xlabel("STATES") 
plt.ylabel("Sec_Only") 
plt.title("Students enrolled in COMPUTER courses") 
plt.xticks(rotation=80)
plt.show()

# creating the bar plot for higher secondary schools
fig = plt.figure(figsize = (20,5)) 
plt.bar(x=comp['State_UT'], height= comp['HrSec_Only'],color ='maroon', width = 0.8) 
plt.xlabel("STATES") 
plt.ylabel("HrSec_Only") 
plt.title("Students enrolled in COMPUTER courses") 
plt.xticks(rotation=80)
plt.show()
```

### Graph-1 shows that in primary section, Union territories (Lakshadweep,Puducherry) and states(Kerala) has contirbuted most towards 'computerised' schools.

### Graph-2 shows that in upper-primary section,  Union territories has contributed most towards 'computerised' schools except Chandigarh(no contibution).

### Graph-3 shows that in secondary section, Union territories (Andaman & Nicobar,Chandigarh,Delhi,Lakshadweep,Puducherry)  and states(Goa,Haryana,Sikkim,Tripura,WestBengal) has no contirbution towards 'computerised' schools.

### Graph-4 shows that in higher-secondary section, Union territories (Andaman & Nicobar,Chandigarh,Lakshadweep)  and   states(Sikkim,Tripura) has no contribution towards 'computerised' schools.

```python
fig = plt.figure(figsize = (20, 5)) 
  
# creating the bar plot 
plt.bar(x=comp['State_UT'], height= comp['All Schools'],color ='green', width = 0.4) 
plt.xlabel("STATES") 
plt.ylabel("ALL SCHOOLS") 
plt.title("Students enrolled in COMPUTER courses") 
plt.xticks(rotation=80)
plt.show() 
```

### Gujarat,Maharashtra and Punjab have also done a pretty good job .
